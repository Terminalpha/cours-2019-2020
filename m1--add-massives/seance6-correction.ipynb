### Agrégat

##### 1. Quelles sont les 10 plus grandes chaines de restaurants (nom identique) ?

```python
affiche(db2.restaurants.aggregate([
    { "$match": { "name": { "$ne": "" }}},
    { "$group": { "_id": "$name", "nb_restaurants": { "$sum": 1 } }},
    { "$sort": { "nb_restaurants": -1 }},
    { "$limit": 10 }
]))
```

##### 2. Lister par quartier le nombre de restaurants et le score moyen

```python
affiche(db2.restaurants.aggregate([
    { "$match": { "borough": { "$ne": "Missing" }}},
    { "$unwind": "$grades" },
    { "$group": { 
        "_id": "$borough", 
        "nb_restaurants": { "$sum": 1 },
        "score_moyen": { "$avg": "$grades.score" }
    }}
]))
```

##### 3. Donner le Top 5 et le Flop 5 des types de cuisine, en terme de nombre de restaurants

```python
# TOP 5
affiche(db2.restaurants.aggregate([
    { "$group": { "_id": "$cuisine", "nb_restaurants": { "$sum": 1 } }},
    { "$sort": { "nb_restaurants": -1 }},
    { "$limit": 5 }
]))
```

```python
# FLOP 5
affiche(db2.restaurants.aggregate([
    { "$group": { "_id": "$cuisine", "nb_restaurants": { "$sum": 1 } }},
    { "$sort": { "nb_restaurants": 1 }},
    { "$limit": 5 }
]))
```

##### 4. Donner les dates de début et de fin des évaluations

```python
affiche(db2.restaurants.aggregate([
    { "$unwind": "$grades" },
    { "$group": { 
        "_id": "dates limites", 
        "debut": { "$min": "$grades.date" },
        "fin": { "$max": "$grades.date" } 
    }}
]))
```

##### 5. Quels sont les 10 restaurants (nom, quartier, addresse et score) avec le plus petit score moyen ?

```python
affiche(db2.restaurants.aggregate([
    { "$unwind": "$grades" },
    { "$group": { 
        "_id": { "nom": "$name", "quartier": "$borough"}, #, "adresse": "$address" }, 
        "score_moyen": { "$avg": "$grades.score" }
    }},
    { "$match" : { "score_moyen": { "$ne": None }}},
    { "$sort": { "score_moyen": 1 }},
    { "$limit": 10 }
]))
```

##### 7. Compter le nombre d’évaluation par jour de la semaine

```python
# Le premier jour de la semaine est le dimanche

affiche(db2.restaurants.aggregate([
    { "$unwind": "$grades" },
    { "$project": { "jour": { "$dayOfWeek": "$grades.date" }}},
    { "$group": { 
        "_id": "$jour", 
        "nb_evaluations": { "$sum": 1 }
    }},
    { "$sort": { "_id": 1 }}
]))
```
